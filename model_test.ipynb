### Carregando o meu modelo treinado.

```python
from ultralytics import YOLO

# Carregar o modelo treinado, Caminho para o modelo salvo
model = YOLO("C:\\Users\\User\\Desktop\\deteccao-rede-YOLO\\runs\\detect\\train\\weights\\best.pt")
```

### Detectando a imagem.

```python
# Caminho da imagem a ser testada
image_path = "C:\\Users\\User\\Downloads\\YOLO_dataset\\imagens_test\\patinete_teste3.jpg"

# Fazer a detecção
results = model(image_path)  # O YOLO já faz a inferência diretamente

# Exibir os resultados
results[0].show()  # Abre a imagem com a detecção desenhada

# Salvar a imagem com os resultados
results[0].save("C:/Users/User/Downloads/YOLO_dataset/imagens_test/patinete_teste3_detectado.png")

```

### Carregando a imagem detectada.

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem detectada
img_detected = cv2.imread("C:/Users/User/Downloads/YOLO_dataset/imagens_test/patinete_teste3_detectado.png")

# Converter de BGR para RGB (o OpenCV usa BGR por padrão)
img_detected = cv2.cvtColor(img_detected, cv2.COLOR_BGR2RGB)

# Mostrar no notebook
plt.figure(figsize=(8, 8))
plt.imshow(img_detected)
plt.axis("off")
plt.show()
```
